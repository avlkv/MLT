# Текстовое описание набора данных
В качестве набора данных мы будем использовать набор данных о заболеваниях сердца.
https://www.kaggle.com/ronitf/heart-disease-uci

Анализ подобного набора данных позволяет определить тенденции в данных о сердце для прогнозирования некоторых сердечно-сосудистых заболеваний или выявить какие-либо четкие признаки состояния сердечно-сосудистого здоровья.

Датасет состоит из одного файла heart.csv .

Файл содержит следующие колонки:

age - возраст в годах<br>
sex - пол (1 - мужской, 0 - женский)<br>
cp - тип боли в груди<br>
trestbps - кровяное давление в спокойном состоянии (в mm Hg при поступлении в больницу)<br>
chol - уровень холестирина в mg/dl<br>
fbs - уровень сахара > 120 mg/dl (1 - да, 0 - нет)<br>
restecg - результат ЭКГ (значения 0, 1, 2)<br>
thal - макс. зафиксированный пульс<br>
exang - ангина, вызванная управжнениями (1 - да, 0 - нет)<br>
oldpeak - снижение ST-зубца, вызванного упражнениями, по сравнению со спокойным состоянием<br>
slope - уклон пикового отрезка ST-зубца<br>
ca - число крупных судов (0-3), окрашенных по цвету<br>
thal - 3 = нормально; 6 - исправленный дефект; 7 - исправляемый дефект<br>
target - целевой признак (0 или 1)<br>

```python
data = pd.read_csv('../data/heart.csv', sep=",")
```

# Проверка на наличие пропусков в данных

```python
# Цикл по колонкам датасета
for col in data.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

# Корреляционный анализ

## Тепловая карта корреляций всех признаков

```python
plt.figure(figsize = (15,10))
sns.heatmap(data.corr(), cmap='coolwarm', annot=True, fmt='.2f')
```

Таким образом, <b>наибольшая корреляция</b> среди всех признаков наблюдается между:
<ul>
    <li>Снижением ST-зубца, вызванного упражнениями, по сравнению с состоянием покоя и уклоном пикового отрезка ST-зубца: -0.58
    <li>Ангиной, вызванной управжнениями и целевым признаком: -0.44</li>
    <li>Типом боли в груди и целевым признаком: 0.43</li>
</ul>

## Список наибольших по модулю корреляций

```python
data.corr().unstack().abs().sort_values(ascending = False).drop_duplicates()[:10]
```

<b>Примечание</b>: отсутствие переменной в левом столбце означает, что там стоит та же переменная, что и в строке выше

## Корреляции целевого признака

```python
plt.figure(figsize = (5,15))
sns.heatmap(data.corr().abs()[['target']].sort_values(by=['target'],ascending=False),cmap='Greens',annot=True);
```

Исходя из данного рисунка можно сделать вывод о том, что определения типа сердечно-сосудистого заболения и вообще его наличия следует прежде всего смотреть на:
<ul>
    <li>наличие ангины, вызванной управжнениями
    <li>тип боли в груди
    <li>снижение ST-зубца, вызванного упражнениями, по сравнению со спокойным состоянием
    </ul>
